* area_m2    : Dairenin yüzölcümü,
* owner      : Dairenin sahibi,
* room       : Oda sayisi,

* date       : Ilan tarihi

```python
df1=df["location"].str.split("/")
loc_list=[]
for i in df1:
    loc_list.append(i[2])
print(loc_list)
```

```python
loc_list1=[]
for y in loc_list:
    loc_list1.append(y.replace("-emlakcidan-daire",""))
    
print(loc_list1)
```

```python
list_ilce=[]
list_semt=[]
for x in loc_list1:
    a=x.split("-")
    list_ilce.append(a[1])
    list_semt.append(a[2])
print(list_ilce)   
    
```
